```python
# Machine Learning Model

# linear regrresion modal
from sklearn.linear_model import LinearRegression

# Metrics / Error Measurement

from sklearn.metrics import mean_squared_error
```

```python
# Pima Women Dataset

# www.google.com/search?client=opera-gx&q=pydataset+documentation&sourceid=opera&ie=UTF-8&oe=UTF-8
pima = data('Pima.tr')  # tr - table relation table
```

```python
# npgreg (number of pregnancy)
# glu (glucose)
# bp (blood pressure)
# skin (thickness of skin'fats)
# bmi
# ped (performance enhancing drug)
# age
# type (I diabetes or not)
```

```python
x = pima['skin']
y = pima['bmi']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel('Skin Thickness')
plt.ylabel('B.M.I')

# Strong positive -> as 'x' increase, 'y' also increase
# Strong negative -> vice versa
```

# **Step 2: Train Test Split**

```python
x.head()
```

```python
y.head()
```

```python
x_test, x_train, y_test, y_train = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

train modal

```python
model = LinearRegression()

# training the model

model.fix(x_train.to_frame(), y_train)
```

```python
y_pred =  model.predict(x_test.to_frame())
```

```python
y_pred
```

```python
mse_error = mean_squared_error(y_pred, y_test)
print(f'MSE Error {mse_error}')
```
