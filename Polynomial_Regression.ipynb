```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

# import model
# linear then only polynomial
from sklearn.linear_model import LinearRegression

# preposessing -> transform data into more suitable for the model
from sklearn.preprocessing import PolynomialFeatures

# to check accuracy of data
from sklearn.metrics import mean_squared_error # OR mean_absolute_error OR root_mean_squared_error
```

```python
# Car Vs Age
np.random.seed(42)

#sample Data
x = np.arange(1,11).reshape(-1,1)
```

```python
# using for-loop to generate 4 BEST FIT LINE with 4 different degrees
```

```python
y = 100 - (x**2) + np.random.normal(0,10, x.shape[0]).reshape(-1,1)
```

```python
# Visualizing the raw data
plt.scatter(x,y, color = 'red')
plt.xlabel('Age')
plt.ylabel('Car Price')
plt.title('Car Price vs Age')

# Using For Loop to generate 4 best fit lines with 4 different degrees
for degree in range(1,5): # [1,2,3,4]
  # Step 1: Transform the original feature to have polynomial feature
  poly = PolynomialFeatures(degree=degree)
  X_poly = poly.fit_transform(x) # Now your X have polynomial features, whichs allow to bend
  # Step 2: Create a Linear Regression Model
  poly_model = LinearRegression()
  poly_model.fit(X_poly,y) # Fit the polynomial data
  # Step 3: Make prediction
  y_pred = poly_model.predict(X_poly)
  # Step 4: Calculating the RMSE
  mse = mean_squared_error(y_pred, y)
  rmse = np.sqrt(mse)
  # Step 5: Plotting the polynomial lines to the graph
  plt.plot(x, y_pred, label = f"Degree {degree}, RMSE {rmse : .2f}")

plt.legend()
plt.show()
```
